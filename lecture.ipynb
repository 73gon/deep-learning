```python
import numpy as np
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

print(sigmoid(-3), sigmoid(2), sigmoid(0), sigmoid(7))

res = 3 * sigmoid(-3) + -12 * sigmoid(2) + sigmoid(0) * 8 + sigmoid(7) + 6

print('result: ', res)
print('sigmoid of result: ', sigmoid(res))
```

```python
def softmax(x):
  return np.exp(x) / np.sum(np.exp(x))

f_in_1 = np.array([1, 5, 0])
f_in_2 = np.array([1.0, 1.1, 0.9])

print(softmax(f_in_1))
print(softmax(f_in_2))
```
